```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the file
file_path = "C:/Users/abomi/Desktop/TFM OK/TFM-main/datas/Brazil_Energy balance matrix.xlsx"

#File excel with multi-sheet_names is read as an Dictionary
Dict = pd.read_excel(file_path, sheet_name=None, skiprows=range(4), skipfooter=3)

for k in Dict.keys():       
    # Round the values of each column in dataframe with 2 decimals
    Dict[k]=Dict[k].applymap(lambda x: round(x, 2) if isinstance(x, (int, float)) else x)
    
    #Ignor the first row (unit) in each df
    Dict[k]=Dict[k][1:]
    
    #Rename of columns in df
    Dict[k].rename(columns={'Unnamed: 0': 'SECTOR','OTHER PRIMARY_x000d_\n':'OTHER PRIMARY'}, inplace=True)   
    
print(Dict.keys())

# Convert the keys of Dict to only contain the year
for k in list(Dict.keys()):
    k_new = k.split(' - ')[0]
    Dict[k_new] = Dict.pop(k)
    Dict[k_new]["Year"]=k_new
        
print('\n')
print(Dict.keys())



#Save the Clean Dictionary in Excel with multi-sheet_names
file_path="C:/Users/abomi/Desktop/TFM OK/TFM-main/datas/Brazil_Energy balance matrix_cleaned.xlsx"
with pd.ExcelWriter(file_path) as writer:
    for k, df in Dict.items():
        df.to_excel(writer, sheet_name=k)

df_2021=Dict['2021']
df_2021.tail(14)
```

```python
Dict2=Dict.copy()

for k in Dict2.keys():
   # selected_columns = ['SECTOR', 'OIL', 'NATURAL GAS', 'COAL', 'HYDROENERGY', 'GEOTHERMAL',
    #   'NUCLEAR', 'FIREWOOD', 'SUGARCANE AND PRODUCTS', 'OTHER PRIMARY', 'TOTAL PRIMARIES','Year']
    selected_columns = ['SECTOR','DIESEL OIL','TOTAL SECUNDARIES','Year']
    Dict2[k]=Dict2[k][selected_columns]
    selected_rows= ['   TRANSPORT ',
       'FINAL CONSUMPTION']
    Dict2[k]=Dict2[k][Dict2[k]['SECTOR'].isin(selected_rows)]
    
    # Concatenate all DataFrames in Dict vertically
    df_conc = pd.concat(Dict2.values(), ignore_index=True)
    
    # Move the column "Year" to the first position
    last_column_name = df_conc.columns[-1]
    # Reorder the columns by concatenating the last column with the rest of the DataFrame
    df_conc = pd.concat([df_conc[last_column_name], df_conc.drop(columns=last_column_name)], axis=1) # Datafames of TOTAL SUPPLY
    df_conc.set_index("Year", inplace=True)
    df_conc.index = df_conc.index.astype(int)
    #df_TC1=df_conc.drop(columns="SECTOR")
    # Set the value "NaN" by Zero
    df_TC1 = df_conc.apply(lambda column: column.fillna(0))
    

df_TC1
```

```python
# Group by Year and SECTOR, and sum the GASOLINE/ALCOHOL and TOTAL SECONDARY values
grouped_df = df_TC1[df_TC1['SECTOR'] != 'FINAL CONSUMPTION'].groupby(['Year', 'SECTOR'])[['DIESEL OIL', 'TOTAL SECUNDARIES']].sum().reset_index()

# Pivot the data
pivot_df = grouped_df.pivot(index='Year', columns='SECTOR', values=['DIESEL OIL', 'TOTAL SECUNDARIES'])

# Display the result
pivot_df.head()

```

```python
# Pivot the data
pivot_df = grouped_df.pivot(index='Year', columns='SECTOR')
pivot_df.columns = pivot_df.columns.map(lambda x: x[0].strip())
```

```python
pivot_df.columns = pivot_df.columns.str.strip()
```

```python
pivot_df['DIESEL OIL/TS'] = pivot_df['DIESEL OIL'] / pivot_df['TOTAL SECUNDARIES']
```

```python

# Cargar los datos económicos y demográficos desde el archivo Excel
economic_demographic_data = pd.read_excel('C:/Users/abomi/Desktop/TFM OK/Mi trabajo/word introduccion + indices/Brasil_indices.xlsx', index_col='Year')

# Unir los DataFrames pivot_df y economic_demographic_data usando el índice (Year)
merged_df = pivot_df.merge(economic_demographic_data, left_index=True, right_index=True)

# Imprimir el DataFrame unido para verificar
print(merged_df)
```

```python
# Verificar valores faltantes
missing_values = merged_df.isnull().sum()
print("Valores faltantes por columna:")
print(missing_values)
```

```python
# Verificar si los datos son numéricos
non_numeric_columns = merged_df.select_dtypes(exclude=[np.number]).columns
print("Columnas no numéricas:")
print(non_numeric_columns)
```

```python
# Convertir la columna 'Población' a numérica en el DataFrame merged_df
merged_df['Población'] = pd.to_numeric(merged_df['Población'], errors='coerce')

# Verificar el tipo de dato después de la conversión
print(merged_df.dtypes)
```

```python
columns_to_drop = ['Tasas de muerte/1000 personas','tasa de crecimiento', 'Consumo per capita energetico', 'Peso Petróleo', 'Peso Natural gas', 
                   'Peso Coal', 'Peso Hydroenergy', 'Nuclear', 'Firewood', 'Sugarcane and products', 'Otros', 'TOTAL', 
                   ' Penet_energ_renovable', 'Huella_carbono', 'Crecimiento población anual']

merged_df.drop(columns=columns_to_drop, inplace=True)

# Verificar el DataFrame después de eliminar las columnas
print(merged_df.head())
```

```python
# Escalar las variables independientes
columns_to_scale = [ 'TOTAL SECUNDARIES', 'Población', 'PIB',
                    'Tasa de natalidad', 'Tasa inflacion']

mean_values = merged_df[columns_to_scale].mean()
std_values = merged_df[columns_to_scale].std()

scaled_data = (merged_df[columns_to_scale] - mean_values) / std_values

scaled_df = pd.concat([scaled_data, merged_df.drop(columns=columns_to_scale)], axis=1)

# Definir las variables de entrada (X) y la variable objetivo (y)
columnas_a_eliminar = ['DIESEL OIL/TS', 'DIESEL OIL']
X = scaled_df.drop(columns=columnas_a_eliminar)
y = scaled_df['DIESEL OIL/TS']

# Dividir los datos en entrenamiento y prueba
train_mask = (scaled_df.index <= 2010)
test_mask = (scaled_df.index >= 2011)

X_train = X[train_mask]
y_train = y[train_mask]

X_test = X[test_mask]
y_test = y[test_mask]

# Imprimir las dimensiones de los conjuntos de entrenamiento y prueba
print("Dimensiones del conjunto de entrenamiento:")
print("X_train:", X_train.shape)
print("y_train:", y_train.shape)

print("\nDimensiones del conjunto de prueba:")
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Crear una instancia del modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Realizar predicciones en los datos de prueba
y_pred = model.predict(X_test)

# Comparar las predicciones con los valores reales usando el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

```python
# Filtrar datos reales y predichos para los años 2011 a 2021
actual_y_2011_2021 = merged_df.loc[(merged_df.index >= 2011) & (merged_df.index <= 2021), 'DIESEL OIL/TS']
predicted_y_2011_2021 = y_pred[(X_test.index >= 2011) & (X_test.index <= 2021)]

# Calcular el error cuadrático medio (MSE) para los años 2011 a 2021
mse_2011_2021 = mean_squared_error(actual_y_2011_2021, predicted_y_2011_2021)

# Calcular el error para cada año en el período 2011-2021
for year in range(2011, 2022):
    actual_value = actual_y_2011_2021[year]
    predicted_value = predicted_y_2011_2021[year - 2011]
    error = actual_value - predicted_value
    print(f"Año {year}: Valor real = {actual_value:.2f}, Valor predicho = {predicted_value:.2f}, Error = {error:.2f}")

print("Mean Squared Error (2011-2021):", mse_2011_2021)
```
